```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
from sklearn import datasets
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/avynychenko/ML-course-Python/master/Datasets%20for%20ML%20course/svm_data.csv', header = None)
data.columns = ['Target', 'Column1', 'Column2']
y = data.Target
X_data = data.drop(['Target'], axis=1)
y.shape
```

```python
## Visualizing data depending on groups (1, 0)
sns.scatterplot(x = 'Column1', y = 'Column2', hue = 'Target', data = data)
```

```python
## Такое значение параметра 'C' нужно использовать, чтобы убедиться, что SVM работает с выборкой как с линейно разделимой.
clf = svm.SVC(C = 100000, random_state=241, kernel = 'linear')
clf.fit(X_data, y)
```

```python
plot_decision_regions(X=np.array(X_data), 
                      y=np.array(y).astype(np.integer),
                      clf=clf, 
                      legend=2)
```

### SVM for NON-Linear Data

```python
circles, target = datasets.make_circles(300, noise=0.1, factor = 0.75)
```

```python
sns.scatterplot(x = circles[:,0], y = circles[:,1], hue = target)
```

```python
clf = svm.SVC(kernel = 'rbf')
params = {'C': [0.01, 0.1, 1, 10, 100],
          'gamma': [0.1, 1, 3, 10, 100]}
cv = GridSearchCV(clf, param_grid = params, cv = 5)
cv.fit(circles, target)
```

```python
SVM = cv.best_estimator_
SVM
```

```python
SVM.fit(circles, target)
plot_decision_regions(X=circles, 
                      y=target,
                      clf=SVM, 
                      legend=2)
```
