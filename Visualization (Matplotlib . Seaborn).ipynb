```python
%matplotlib inline
import matplotlib as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

Библиотека seaborn - это надстройка над matplotlib, которая позволяет строить более симпатичные графики.

```python
students_performance = pd.read_csv('Datasets for ML course/StudentsPerformance.csv')
students_performance = students_performance.rename(columns = {'parental level of education':'parental_level_of_education',
                                                             'test preparation course': 'test_preparation_course',
                                                             'math score': 'math_score',
                                                             'reading score': 'reading_score',
                                                             'writing score': 'writing_score'})
```

**Построить гистограмму** по встроенной функции по определенным переменным:

```python
students_performance.math_score.hist()
```

**Нарисовать SCATTER PLOT** для 2-х переменных

```python
students_performance.plot.scatter(x = 'math_score', y = 'reading_score')
```

В функции lmplot от seaborn есть много аргументов, которые позволяют менять настройки графика. Если мы хотим:
    - убрать регрессионные прямые, то можно использовать аргумент "fit_reg = False"
    - сгруппировать данные по переменной "hue = '*название переменной*' "

```python
plot = sns.lmplot(x = 'math_score', y = 'reading_score', data = students_performance, hue = 'gender', palette = sns.color_palette("husl"))
```

 Также можно пименять различные методы к графику, чтобы например **изменить название осей, название графика и т.д.**

```python
plot.set_xlabels('Math score')
plot.set_ylabels('Reading score')
# plot.set_titles('Correlation between reading scores and math scores')
plot.fig
```

**KDE GRAPHS** 

```python
iris = pd.read_csv('/Users/avyny/ML_course_tasks(Python)/Datasets for ML course/iris.csv', index_col = 0)
iris = iris.rename(columns = {'sepal length': 'sepal_length',
                             'sepal width': 'sepal_width',
                             'petal length': 'petal_length',
                             'petal width': 'petal_width'})
iris_without_species = iris.drop(['species'], axis = 1)
iris_without_species.apply(lambda x: sns.kdeplot(x))
```

```python
sns.violinplot(y = 'petal_length', data = iris_without_species);
```

**PAIRPLOT GRAPH**

```python
iris['species'] = iris['species'].astype(object)
sns.pairplot(data = iris, hue = 'species');
```
